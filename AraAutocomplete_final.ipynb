# **Preprocessing**

```python
from datasets import load_dataset

# Load Arabic part of Oscar dataset
dataset = load_dataset("oscar", "unshuffled_deduplicated_ar", split="train[:1%]")
# Using only 1% for quick training
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "aubmindlab/aragpt2-medium"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

tokenizer.pad_token = tokenizer.eos_token
model.config.pad_token_id = tokenizer.pad_token_id
```

```python

def tokenize_function(examples):
    return tokenizer(examples["text"], truncation=True, max_length=128)

tokenized_dataset = dataset.map(tokenize_function, batched=True, remove_columns=["text"])
```

```python
split_dataset = tokenized_dataset.train_test_split(test_size=0.1)
eval_dataset = split_dataset["test"]
```

# **Training**

```python
from transformers import Trainer, TrainingArguments, DataCollatorForLanguageModeling

# Automatically mask tokens randomly (not using MLM for causal language modeling)
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer, mlm=False
)

# ✅ Set your checkpoint directory
checkpoint_dir = "/content/second_drive/MyDrive/arabic_autocomplete_checkpoints_31000"

training_args = TrainingArguments(
    output_dir="./arabic_autocomplete",
    per_device_train_batch_size=4,
    learning_rate=5e-5,
    num_train_epochs=2,
    weight_decay=0.01,
    save_total_limit=2,               # only keep last 2 checkpoints
    save_steps=500,                   # save every 500 steps
    save_strategy="steps",            # save based on steps
    logging_dir="./logs",             # for TensorBoard if you want
    logging_steps=100,                # optional: log training loss every 100 steps
    overwrite_output_dir=True         # ✅ important if reusing the same output_dir
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=data_collator,
)

# ✅ Train from checkpoint
trainer.train(resume_from_checkpoint=checkpoint_dir)
```

# **Checkpoint checking**

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

# Set checkpoint path
checkpoint_path = "/content/second_drive/MyDrive/arabic_autocomplete_checkpoints_45036/"  # ← Replace XXXX with the actual step number

# Load tokenizer from base model (same one you used originally)
tokenizer = AutoTokenizer.from_pretrained("aubmindlab/aragpt2-medium")  # or "gpt2", etc.

# Load trained model from checkpoint
model = AutoModelForCausalLM.from_pretrained(checkpoint_path)

# Fix pad_token if needed
tokenizer.pad_token = tokenizer.eos_token
model.config.pad_token_id = tokenizer.pad_token_id

# Set up text generation pipeline
text_generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

# Example Arabic prompt
prompt = "ما هو"

# Generate text
output = text_generator(
    prompt,
    max_length=100,
    num_return_sequences=1,
    do_sample=True,
    top_k=50,
    top_p=0.95,
    temperature=0.9,
    repetition_penalty=1.2,
)

# Print result
print("📝 Generated Arabic Text:\n")
print(output[0]["generated_text"])
```

# **Evaluation**

```python
from sklearn.metrics import accuracy_score
import numpy as np

def compute_metrics(eval_preds):
    logits, labels = eval_preds
    predictions = np.argmax(logits, axis=-1)

    # Mask padding tokens
    mask = labels != -100
    correct = (predictions == labels) & mask
    accuracy = correct.sum() / mask.sum()

    # Cross entropy loss is already handled by Trainer
    return {
        "eval_accuracy": accuracy,
    }

# Reinitialize Trainer with compute_metrics
trainer = Trainer(
    model=model,
    args=training_args,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)

metrics = trainer.evaluate(eval_dataset=eval_dataset.select(range(90)))
print(metrics)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import gradio as gr
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
```

```python
from transformers import GPT2Config, GPT2LMHeadModel
from safetensors import safe_open
from transformers import AutoModelForCausalLM, AutoTokenizer


config = GPT2Config(
    vocab_size=64000,        # Matches wte.weight shape [64000, 1024]
    n_positions=1024,        # Context length
    n_embd=1024,            # Hidden size (matches wte/wpe weight dim 1)
    n_layer=24,              # 24 transformer blocks (h.0 to h.23)
    n_head=16,               # 1024/16 = 64 head dimension
    n_inner=4096,            # MLP intermediate size (matches c_fc.weight[1])
    bos_token_id=0,
    eos_token_id=2,
    pad_token_id=1
)

model_path = "/content/second_drive/MyDrive/arabic_autocomplete_checkpoints_45036/"
# 1. Load tokenizer first
tokenizer = AutoTokenizer.from_pretrained(model_path)
# 2. Load model with correct config
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    config=config,
    use_safetensors=True
)
```

```python
# Set pad token if missing
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

def generate_text(input_text, length=30):
    if not input_text:
        return ""

    try:
        # Tokenize with attention mask
        inputs = tokenizer(
            input_text,
            return_tensors='pt',
            padding=True,
            truncation=True
        )

        # Generate continuation
        outputs = model.generate(
            inputs.input_ids,
            attention_mask=inputs.attention_mask,
            max_length=inputs.input_ids.shape[-1] + length,
            do_sample=True,
            top_k=50,
            top_p=0.95,
            temperature=0.9,
            pad_token_id=tokenizer.eos_token_id
        )

        # Return full generated text (input + completion)
        full_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
        return full_text.strip()

    except Exception as e:
        return f"Generation Error: {str(e)}"

# Arabic interface setup
interface = gr.Interface(
    fn=generate_text,
    inputs=[
        gr.Textbox(label="النص المدخل", lines=5),
        gr.Slider(10, 100, value=30, label="طول الإكمال")
    ],
    outputs=gr.Textbox(label="النص الكامل", lines=10),
    title="الإكمال التلقائي للغة العربية",
    description="أدخل النص العربي وسيتم إكماله مع عرض النص الأصلي:",
    allow_flagging="never"
)

interface.launch(debug=True)
```
